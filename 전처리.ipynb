# 날짜 컬럼 추가
year, month, dayofweek, year-month

```python
#year컬럼추가
df["year"] = df["transacted_date"].dt.year
df.head(1)
```

```python
#month컬럼추가
df["month"] = df["transacted_date"].dt.month
df.head(1)
```

```python
#dayofweek컬럼추가
df["dayofweek"] = df["transacted_date"].dt.dayofweek
df.head(1)
```

```python
df[["year","month"]].astype('object')
```

```python
#year-month컬럼추가
df[["year","month"]]=df[["year","month"]].astype('str')
df['year-month'] = df['year'].str.cat(df['month'],sep="-")
df[["year","month"]]=df[["year","month"]].astype('int')
df.head(1)
```

# season, quarter 컬럼

```python
#season컬럼추가
#spring:3,4,5 summer:6,7,8 fall: 9,10,11 winter:12,1,2
df.loc[(df["month"]>=3)&(df["month"]<6),"season"] = "spring"
df.loc[(df["month"]>=6)&(df["month"]<9),"season"] = "summer"
df.loc[(df["month"]>=9)&(df["month"]<12),"season"] = "fall"
df.loc[(df["month"]==12),"season"] = "winter"
df.loc[(df["month"]<3),"season"] = "winter"
df.head()
```

```python
#quarter컬럼추가
df.loc[(df["month"]>=1)&(df["month"]<=3),"quarter"] = 1
df.loc[(df["month"]>=4)&(df["month"]<=6),"quarter"] = 2
df.loc[(df["month"]>=7)&(df["month"]<=9),"quarter"] = 3
df.loc[(df["month"]>=10)&(df["month"]<=12),"quarter"] = 4
df.head()
```

```python
df['quarter']=df['quarter'].astype(int)
```

```python
df.info()
```
